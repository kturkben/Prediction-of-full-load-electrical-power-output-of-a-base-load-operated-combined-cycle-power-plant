```python
corr=df.corr()
corr
```

```python
import matplotlib.pyplot as plt
import seaborn
plt.figure(figsize=(12, 8))
heatmap = seaborn.heatmap(corr, vmin=-1, vmax=1, annot=True, cmap='YLGnBu')
heatmap.set_title('Correlation Heatmap');
```

```python
for col in df:
    maxD=abs(corr[col]).nlargest(n=2)
    #print(maxD)
    for index,value in maxD.items():
        if 1>value>0.70:
            print(index,col,'değişkenleri yüksek korelasyona sahip',value)
```

LINEER REGRESSION MODEL KURULUMU

```python
X = df.iloc[:,[0,1,2,3,4,6,7,8]].values
y = df.iloc[:, 5].values#bağımlı d.
#print(df)
print(X)
```

MinMaxScaler() İLE NORMALİZASYON

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
scaled_df = min_max_scaler.fit_transform(df)
df_sc = pd.DataFrame(scaled_df)
print(df_sc)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 0)#the best
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import math
model=LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(model.coef_.round(3))
print(model.intercept_.round(3))
mse=mean_squared_error (y_test,y_pred)
print('MAE : ',mean_absolute_error (y_test,y_pred))
print('MSE : ',mean_squared_error (y_test,y_pred))
print('RMSE: ', math. sqrt(mse))
print('R^2 Score: ',r2_score(y_test,y_pred,force_finite=False))
#RMSE değerinin sıfır olması modelin hiç hata yapmadığı anlamına gelir
#mse 0'a yakın olmalı
plt.plot(y_test, label='Gerçek')
plt.plot(y_pred, label='Tahmin')
plt.legend()
```

```python
#yüksek korelasyonlu iki değişken çıkarıldıktan sonraki sonuçlar:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 0)#the best
model=LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(model.coef_.round(3))
print(model.intercept_.round(3))
mse=mean_squared_error (y_test,y_pred)
print('MAE : ',mean_absolute_error (y_test,y_pred))
print('MSE : ',mean_squared_error (y_test,y_pred))
print('RMSE: ', math. sqrt(mse))
print('R^2 Score: ',r2_score(y_test,y_pred,force_finite=False))
plt.plot(y_test, label='Gerçek')
plt.plot(y_pred, label='Tahmin')
plt.legend()
```
